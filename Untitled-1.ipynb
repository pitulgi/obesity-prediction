```python
import pandas as pd
import plotly.express as px
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
import seaborn as sns
from sklearn.ensemble import RandomForestClassifier
from sklearn.naive_bayes import GaussianNB
from sklearn.linear_model import LogisticRegression
from xgboost import XGBClassifier, plot_importance
from sklearn.preprocessing import LabelEncoder, StandardScaler
from sklearn.pipeline import Pipeline
import pickle
```

```python
from ucimlrepo import fetch_ucirepo 
  
# fetch dataset 
estimation_of_obesity_levels_based_on_eating_habits_and_physical_condition = fetch_ucirepo(id=544) 
  
# data (as pandas dataframes) 
X = estimation_of_obesity_levels_based_on_eating_habits_and_physical_condition.data.features 
y = estimation_of_obesity_levels_based_on_eating_habits_and_physical_condition.data.targets 
  
# metadata 
print(estimation_of_obesity_levels_based_on_eating_habits_and_physical_condition.metadata) 
  
# variable information 
print(estimation_of_obesity_levels_based_on_eating_habits_and_physical_condition.variables) 

X.head(10)
```

```python
X = pd.DataFrame(X)  # если ещё не DataFrame
#y = pd.Series(y, name='Obesity_Level')  # задаём имя целевому столбцу
df_full = pd.concat([X, y], axis=1)
df_full = df_full.rename(columns={'family_history_with_overweight': 'FHWO'})

encoding_map = {
    'Gender': {'Male': 0, 'Female': 1},
    'FHWO': {'no': 0, 'yes': 1},
    'FAVC': {'no': 0, 'yes': 1},
    'SMOKE': {'no': 0, 'yes': 1},
    'SCC': {'no': 0, 'yes': 1},
    'CAEC': {'no': 0, 'Sometimes': 1, 'Frequently': 2, 'Always': 3},
    'CALC': {'no': 0, 'Sometimes': 1, 'Frequently': 2, 'Always': 3},
    'MTRANS': {'Automobile': 0, 'Motorbike': 1, 'Public_Transportation': 2, 'Bike': 3, 'Walking': 4},
    'NObeyesdad': {
        'Insufficient_Weight': 0,
        'Normal_Weight': 1,
        'Overweight_Level_I': 2,
        'Overweight_Level_II': 3,
        'Obesity_Type_I': 4,
        'Obesity_Type_II': 5,
        'Obesity_Type_III': 6
    }
}

df_num = df_full.replace(encoding_map)
df_num.head(20)
```

```python
correlation_matrix = df_num.corr()

rounded_matrix = correlation_matrix.round(2)

fig = px.imshow(
    rounded_matrix,
    text_auto=True, 
    color_continuous_scale='rdbu',
    title='Таблица корреляций',
    aspect="auto"
)

fig.update_layout(
    width=800,
    height=600,
    title_x=0.5,
    xaxis_tickangle=45
)

fig.show()
```

```python
X = df_num.drop('NObeyesdad', axis=1)
y = df_num['NObeyesdad']

X_train, X_test, y_train, y_test = train_test_split(
    X, y, 
    test_size=0.2,      # 80% на обучение, 20% на тест
    random_state=42,    # воспроизводимость
    stratify=y          # сохраняем доли классов
)
```

```python
clf_DTC = DecisionTreeClassifier(
    class_weight='balanced',   # ← ключевая строка
    random_state=42
)

# Обучаем
clf_DTC.fit(X_train, y_train)

y_pred = clf_DTC.predict(X_test)

print(f"Accuracy: {accuracy_score(y_test, y_pred):.4f}\n")
print("Classification Report:")
print(classification_report(y_test, y_pred))

# Вычисляем матрицу ошибок
cm_norm = confusion_matrix(y_test, y_pred, normalize='true')  # по строкам (recall)
sns.heatmap(cm_norm, annot=True, fmt='.2f', cmap='Blues')
```

```python
clf_RFC = RandomForestClassifier(
    n_estimators=100,           # количество деревьев
    class_weight='balanced',    # важнее для малых классов
    random_state=42,
    n_jobs=-1                   # использовать все ядра
)

# Обучаем
clf_RFC.fit(X_train, y_train)

# Предсказания
y_pred = clf_RFC.predict(X_test)

# Точность
print(f"Accuracy: {clf_RFC.score(X_test, y_test):.4f}\n")

# Подробный отчёт
print("Classification Report:")
print(classification_report(y_test, y_pred))

# Вычисляем матрицу ошибок
cm_norm = confusion_matrix(y_test, y_pred, normalize='true')  # по строкам (recall)
sns.heatmap(cm_norm, annot=True, fmt='.2f', cmap='Blues')
```

```python
clf_GNB = GaussianNB()
clf_GNB.fit(X_train, y_train)

# Предсказания
y_pred = clf_GNB.predict(X_test)

# Точность
print(f"Accuracy: {accuracy_score(y_test, y_pred):.4f}\n")

# Отчёт
print("Classification Report:")
print(classification_report(y_test, y_pred))

cm_norm = confusion_matrix(y_test, y_pred, normalize='true')  # по строкам (recall)
sns.heatmap(cm_norm, annot=True, fmt='.2f', cmap='Blues')
```

```python
clf_LR = Pipeline([
    ('scaler', StandardScaler()),  # обязательная нормализация
    ('logreg', LogisticRegression(
        max_iter=3000,              # ↑ увеличили лимит итераций
        solver='saga',              # ✅ лучше для многоклассовой задачи
        class_weight='balanced',    # борьба с дисбалансом
        random_state=42,
        # multi_class='multinomial' — больше не нужно указывать!
        n_jobs=-1                   # ускорение
    ))
])

# Обучаем
clf_LR.fit(X_train, y_train)

# Предсказания
y_pred = clf_LR.predict(X_test)

# Точность
acc = accuracy_score(y_test, y_pred)
print(f"Accuracy: {acc:.4f}\n")

# Отчёт
print("Classification Report:")
print(classification_report(y_test, y_pred))

cm_norm = confusion_matrix(y_test, y_pred, normalize='true')  # по строкам (recall)
sns.heatmap(cm_norm, annot=True, fmt='.2f', cmap='Blues')
```

```python
le_y = LabelEncoder()
y_encoded = le_y.fit_transform(y)

clf_XGB = XGBClassifier(
    objective='multi:softprob',        # для многоклассовой классификации
    eval_metric='mlogloss',            # метрика для отслеживания
    n_estimators=100,
    random_state=42,
    importance_type="gain",
    learning_rate=0.1,
    max_depth=6
)

# Обучаем
clf_XGB.fit(X_train, y_train)

# Предсказания
y_pred = clf_XGB.predict(X_test)

# Для отчёта: верни оригинальные названия классов
y_test_labels = le_y.inverse_transform(y_test)
y_pred_labels = le_y.inverse_transform(y_pred)

# Точность
acc = accuracy_score(y_test, y_pred)
print(f"Accuracy: {acc:.4f}\n")

# Отчёт с оригинальными именами
print("Classification Report:")
print(classification_report(y_test_labels, y_pred_labels))


cm_norm = confusion_matrix(y_test, y_pred, normalize='true')  # по строкам (recall)
sns.heatmap(cm_norm, annot=True, fmt='.2f', cmap='Blues')
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import numpy as np
from sklearn.metrics import (accuracy_score, f1_score, precision_score, 
                            recall_score, roc_auc_score, confusion_matrix, 
                            ConfusionMatrixDisplay)
from sklearn.preprocessing import label_binarize
from itertools import cycle

# Ваши модели (предполагаем X_test, y_test существуют)
models = {
    'Decision Tree': clf_DTC,
    'Random Forest': clf_RFC, 
    'Naive Bayes': clf_GNB,
    'Logistic Reg': clf_LR,
    'XGBoost': clf_XGB
}

# 1. Вычисление метрик (без изменений)
metrics = {}
for name, model in models.items():
    y_pred = model.predict(X_test)
    y_pred_proba = model.predict_proba(X_test) if hasattr(model, 'predict_proba') else None
    
    metrics[name] = {
        'Accuracy': accuracy_score(y_test, y_pred),
        'F1': f1_score(y_test, y_pred, average='weighted'),
        'Precision': precision_score(y_test, y_pred, average='weighted'),
        'Recall': recall_score(y_test, y_pred, average='weighted'),
        'ROC-AUC': roc_auc_score(label_binarize(y_test, classes=model.classes_), 
                               y_pred_proba, multi_class='ovr') if y_pred_proba is not None else np.nan
    }

# ПРАВИЛЬНО: metrics_df с колонками-метриками, строками-моделями
df_metrics = pd.DataFrame(metrics).T.round(3)
print("Таблица метрик:")
print(df_metrics)
```

```python
import numpy as np
import pandas as pd
from sklearn.inspection import permutation_importance

# предполагаем, что clf_RFC, clf_XGB, X_test, y_test, feature_names уже есть
# ВАЖНО: clf_XGB должен быть создан с importance_type="total_gain" или "gain", например:
# clf_XGB = XGBClassifier(..., importance_type="total_gain")

# 1. Random Forest MDI
rf_mdi = clf_RFC.feature_importances_

# 2. XGBoost gain-based importance (из feature_importances_)
xgb_gain = clf_XGB.feature_importances_   # длина = len(feature_names)

# 3. Permutation importance для XGBoost
perm_xgb = permutation_importance(
    clf_XGB, X_test, y_test,
    n_repeats=20,
    random_state=42,
    n_jobs=-1
)
perm_mean = perm_xgb.importances_mean

# 4. Собираем всё в один DataFrame
fi_df = pd.DataFrame({
    "feature": feature_names,
    "rf_mdi": rf_mdi,
    "xgb_gain": xgb_gain,
    "perm_xgb": perm_mean
})

# 5. Ранги для удобного сравнения
for col in ["rf_mdi", "xgb_gain", "perm_xgb"]:
    fi_df[col + "_rank"] = fi_df[col].rank(ascending=False, method="min")

# 6. Средний ранг и сортировка
fi_df["mean_rank"] = fi_df[["rf_mdi_rank", "xgb_gain_rank", "perm_xgb_rank"]].mean(axis=1)
fi_df = fi_df.sort_values("mean_rank")

print(fi_df.head(20))
```

```python
from sklearn.inspection import PartialDependenceDisplay

features_to_plot = ["FCVC", "FAVC", "CH2O"]  # имена должны быть колонками X_test
class_id = 3
  # индекс нужного класса (0..6)

fig, ax = plt.subplots(figsize=(10, 6))

PartialDependenceDisplay.from_estimator(
    clf_XGB,
    X_test,
    features=features_to_plot,
    kind="both",          # 'average', 'individual' или 'both'
    subsample=1000,
    random_state=42,
    target=class_id,      # <--- важно для мультикласса
    ax=ax
)

plt.suptitle(f"PDP и ICE для признаков (класс {class_id})")
plt.tight_layout()
plt.show()
```

```python
import shap
import numpy as np

shap.initjs()  # для интерактивных графиков в ноутбуке

# 1. Эксплейнер и SHAP-значения
explainer_xgb = shap.TreeExplainer(clf_XGB)
shap_values_xgb = explainer_xgb.shap_values(X_test)
# у тебя: (n_samples, n_features, n_classes) = (423, 16, 7)

print("shap_values_xgb shape:", np.array(shap_values_xgb).shape)

# 2. Глобальная важность по всем классам (барплот)
sv_global = np.mean(np.abs(shap_values_xgb), axis=2)  # -> (423, 16)

shap.summary_plot(
    sv_global,
    X_test,
    feature_names=feature_names,
    plot_type="bar",
    show=True
)

# 3. Beeswarm по всем классам (тот же sv_global)
shap.summary_plot(
    sv_global,
    X_test,
    feature_names=feature_names,
    show=True
)

# 4. Локальное объяснение для одного класса и объекта
class_id = 0  # класс 0..6
i = 0         # индекс наблюдения

sv_row = shap_values_xgb[i, :, class_id]  # (16,)
base_value = explainer_xgb.expected_value[class_id]

shap.force_plot(
    base_value,
    sv_row,
    X_test.iloc[i, :],
    matplotlib=True
)
```
